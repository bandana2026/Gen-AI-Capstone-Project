```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
data = pd.read_csv("AusApparalSales4thQrt2020.csv")
```

```python
data.groupby("State")["Sales"].sum()
```

```python
data.groupby("State")["Sales"].sum().plot(kind="bar")
```

```python
data.groupby("State")["Sales"].sum().plot(kind="pie")
```

```python
data.groupby("State")["Sales"].sum().plot(kind="bar",
figsize=(8,5),
title="Total Sales by State")
plt.ylabel("Sales AMount")
plt.show()
```

```python
data.groupby("State")["Sales"].sum().plot(kind="bar", 
figsize=(8,5), 
color="orange", 
title="Total Sales by State")

plt.xlabel("State")
plt.ylabel("Sales Amount")
plt.show()
```

```python
sales_data = data.groupby("State")["Sales"].sum().reset_index()

sns.barplot(
    x="State",
    y="Sales",
    data=sales_data
)

plt.title("State-wise Sales Analysis")
plt.show()
```

```python
sales_sorted = data.groupby("State")["Sales"].sum().sort_values(ascending=False)

sales_sorted.plot(
    kind="bar",
    figsize=(8,5),
    color="green",
    title="State-wise Sales Ranking" 
)

plt.xlabel("State")
plt.ylabel("Total Sales")
plt.show()
```

```python
sales_sorted.plot(
    kind="barh",
    figsize=(8,5),
    color="purple",
    title="Horizontal State-wise Sales Ranking" )

plt.xlabel("Total Sales")
plt.ylabel("State")
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(
    data["Sales"],
    bins=10,
    color="skyblue",
    edgecolor="red"
)


plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequesncy")
plt.show()

```

```python
print("Findings:")
print("1. VIC generated the highest sales.")
print("2. NSW generated the second highest sales.")
print("3. WA generated the lowest sales.")
print("4. Sales are unevenly distributed across states.")
```
